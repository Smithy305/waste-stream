# 02 — Results Analysis

Load experiment results and produce analysis plots: labelled fraction sweep, noise robustness, pseudo-label quality.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")

RESULTS_DIR = Path("../results")
```

```python
# Load all experiment results into a DataFrame
rows = []
for d in sorted(RESULTS_DIR.iterdir()):
    f = d / "eval_results.json"
    if f.exists():
        with open(f) as fh:
            data = json.load(fh)
        data["experiment"] = d.name
        rows.append(data)

df = pd.DataFrame(rows)
df.head()
```
